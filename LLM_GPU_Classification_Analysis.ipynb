```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Ustawienia wykresów
sns.set(style="whitegrid")
```

### Wczytanie danych z pliku Excel

```python
df = pd.read_excel("GPU_Steam_Klasyfikacja_Tak_Nie.xlsx")
df.head()
```

### Konwersja klasy na kategorię z określoną kolejnością

```python
df["ML Class"] = pd.Categorical(df["ML Class"], 
    categories=["None", "ML only", "LLM Basic", "LLM Pro"], 
    ordered=True)
```

### Liczba kart graficznych w każdej klasie ML/LLM

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x="ML Class", order=["None", "ML only", "LLM Basic", "LLM Pro"])
plt.title("Liczba GPU w każdej klasie ML/LLM")
plt.xlabel("Klasa ML/LLM")
plt.ylabel("Liczba kart")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Eksport oczyszczonych danych

```python
df.to_csv("GPU_Cleaned_Classification.csv", index=False)
```
